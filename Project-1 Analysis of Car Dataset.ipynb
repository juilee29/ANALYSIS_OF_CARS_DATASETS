# Bivariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
E
```

```python
#Separating continuous and categorical columns from the dataset
for i in E.columns:
    if(E[i].dtypes == 'object'):
        sb.boxplot(E[i],E.Width)#Boxplot between categorical and continuous columns
        plt.xlabel(i)
        plt.ylabel('Width')
        plt.show()
    else:
        plt.scatter(E[i],E.Width)# Scatter plot between continuous and continuous columns
        plt.xlabel('Width')
        plt.ylabel(i)
        plt.show()
```

```python
for i in E.columns:
   if(E[i].dtypes=="object"):
    sb.countplot(E.Origin,hue=E[i])#Countplot with hue between categorical and categorical columns
    plt.ylabel(i)
    plt.show()
```
